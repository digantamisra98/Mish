```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 12]
import pandas as pd
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
mish_acc = np.load('mish_caltechc256_shufflenetv2x1_acc.npz')
mish_loss = np.load('mish_caltechc256_shufflenetv2x1_loss.npz')
swish_acc = np.load('swish_caltechc256_shufflenetv2x1_acc.npz')
swish_loss = np.load('swish_caltechc256_shufflenetv2x1_loss.npz')
relu_acc = np.load('relu_caltechc256_shufflenetv2x1_acc.npz')
relu_loss = np.load('relu_caltechc256_shufflenetv2x1_loss.npz')
```

```python
lst = mish_acc.files
for item in lst:
    print(item)
    print(mish_acc[item])
```

```python
mish_acc['x']
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(121)
plt.plot(mish_acc['x'], 'r', label= "Training Accuracy (Mish)")
plt.plot(swish_acc['x'], 'b', label= "Training Accuracy (Swish)")
plt.plot(relu_acc['x'], 'g', label= "Training Accuracy (ReLU)")
plt.plot(mish_acc['y'], 'r--', label= "Test Accuracy (Mish)")
plt.plot(swish_acc['y'], 'b--', label= "Test Accuracy (Swish)")
plt.plot(relu_acc['y'], 'g--', label= "Test Accuracy (ReLU)")
ax.legend(loc='best', fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.title.set_text('Accuracy Stats')
ax.title.set_fontsize(20)
ax.grid()
ax2=fig.add_subplot(122)
plt.plot(mish_loss['x'], 'r', label= "Training Loss (Mish)")
plt.plot(swish_loss['x'], 'b', label= "Training Loss (Swish)")
plt.plot(relu_loss['x'], 'g', label= "Training Loss (ReLU)")
plt.plot(mish_loss['y'], 'r--', label= "Test Loss (Mish)")
plt.plot(swish_loss['y'], 'b--', label= "Test Loss (Swish)")
plt.plot(relu_loss['y'], 'g--', label= "Test Loss (ReLU)")
ax2.legend(loc='best', fontsize=15)
ax2.tick_params(axis='both', which='major', labelsize=15)
ax2.title.set_text('Loss Stats')
ax2.title.set_fontsize(20)
ax2.grid()
plt.savefig("shufflev2c256.png", bbox_inches = 'tight')
plt.show()
```
